# Spectral Clustering


In this exercise, you will implement spectral clustering and try it on diffent kinds of data, for instance:

* N-blob: Random points in $\mathbb{R}^2$ according to N Gaussian distributions with different means;

* Two moons: Random points shaped as two intertwined moons; 

* Point and circle: random points from a concentrated Gaussian point in the middle and a wide circle around it.

The first step after generating the data is to construct a graph and, for that, we need a similarity function. In this exercise we will use the following similarity:

$$
d(x_i, x_j) = \exp\left(-\frac{||x_i-x_j||_2^2}{2\sigma^2}\right)
$$

Using this similarity, we can compute the weight matrix W representing the graph, compute the graph Laplacian L and finally use its eigenvectors for clustering.

## Building similarity graph

```python
def build_similarity_graph(X, var=1.0, eps=0.0, k=0):
    """
    TO BE COMPLETED.

    Computes the similarity matrix for a given dataset of samples. 
    If k=0, builds epsilon graph. 
    Otherwise, builds kNN graph.

    Parameters
    ----------
    X : numpy array
        (n x m) matrix of m-dimensional samples
    var : double 
        the sigma value for the exponential function, already squared
    eps : double
        threshold for epsilon graphs
    k : int
        The number of neighbours k for k-nn. If zero, use epsilon-graph

    Returns
    -------
        W: (n x n) dimensional matrix representing the weight matrix of the graph
    """
    n = X.shape[0]
    W = np.zeros((n, n))

    """
    Build similarity graph, before threshold or kNN
    similarities: (n x n) matrix with similarities between all possible couples of points.
    The similarity function is d(x,y)=exp(-||x-y||^2/(2*var))
    """
  
    similarities = np.zeros((n, n))
    for i in range(X.shape[0]):
            for j in range(i+1):
                w=np.exp(-np.sum((X[i]-X[j])**2)/(2*var))
                similarities[i,j]=w
                similarities[j,i]=w

    # If epsilon graph
    if k == 0:
        """
        compute an epsilon graph from the similarities             
        for each node x_i, an epsilon graph has weights             
        w_ij = d(x_i,x_j) when w_ij >= eps, and 0 otherwise          
        """
        for i in range(X.shape[0]):
            for j in range(i+1):
                w=similarities[i,j]
                if w >= eps :
                    W[i,j]=w
                    W[j,i]=w

    # If kNN graph
    elif k != 0:
        """
        compute a k-nn graph from the similarities                   
        for each node x_i, a k-nn graph has weights                  
        w_ij = d(x_i,x_j) for the k closest nodes to x_i, and 0     
        for all the k-n remaining nodes                              
        Remember to remove self similarity and                       
        make the graph undirected                                    
        """
        for i in range(X.shape[0]):
            indices=np.argsort(similarities[i])[X.shape[0]-k-1:]
            for j in indices :
                W[i,j]=similarities[i,j]
            W[i,i]=0
        


    return W
```

## Computing the Laplacian

```python
def build_laplacian(W, laplacian_normalization='unn'):
    """
    Compute graph Laplacian.

    Parameters
    ----------
    W : numpy array
        Adjacency matrix (n x n)
    laplacian_normalization : str
        String selecting which version of the laplacian matrix to construct.
            'unn':  unnormalized,
            'sym': symmetric normalization
            'rw':  random-walk normalization   

    Returns
    -------
    L: (n x n) dimensional matrix representing the Laplacian of the graph
    """
    L = np.zeros(W.shape)
    D = np.zeros(W.shape)
    D_sym = np.zeros(W.shape)
    D_inv = np.zeros(W.shape)
    I = np.eye(W.shape[0])
    for i in range(W.shape[0]):
        D[i,i]=np.sum(W[i])
        D_inv[i,i]=1/D[i,i]
        D_sym[i,i]=1/np.sqrt(D[i,i])
    
    if laplacian_normalization=='unn':
        L= D-W
    if laplacian_normalization=='sym':
        L= I - D_sym @ W @ D_sym
    if laplacian_normalization=='rw' :
        L = I - D_inv@W
    # for example (unnormalized): L = D - W
    return L
```

## Clustering

```python
from scipy.sparse.linalg import eigs
```

```python
def spectral_clustering(L, chosen_eig_indices=None, num_classes=2):
    """
    Parameters
    ----------
    L : numpy array
        Graph Laplacian (standard or normalized)
    choosen_eig_indices : list or None
        Indices of eigenvectors to use for clustering. 
        If None, use adaptive choice of eigenvectors.
    num_classes : int 
        Number of clusters to compute (defaults to 2)
    
    
    Returns
    -------
    Y : numpy array (num_samples, )
        Cluster assignments
    """

    """
    Use the function scipy.linalg.eig or the function scipy.sparse.linalg.eigs to compute:
    U = (n x n) eigenvector matrix           (sorted)
    E = (n x n) eigenvalue diagonal matrix   (sorted)
    """
    

    """
    compute the clustering assignment from the eigenvectors        
    Y = (n x 1) cluster assignments in [0,1,...,num_classes-1]                   
    """
    
    
    ##################
    V,E = eigs(L, which='SR')
    eigenvectors=np.real(E)
    eigenvalues=np.real(V)
    Y = np.zeros(L.shape[0])
    if chosen_eig_indices is not None:
        Y = KMeans(num_classes).fit_predict(eigenvectors[:,chosen_eig_indices])
    else :
        if (eigenvalues[0]==0 and eigenvalues[1]==0) :
            idx_min=0
        else :
            idx_min = np.argmax(eigenvalues > 1e-7)
            
        Y = KMeans(num_classes).fit_predict(eigenvectors[:,idx_min + np.arange(num_classes - 1)])
    return Y
```

## Parameter sensitivity 

```python
def parameter_sensitivity():
    """
    TO BE COMPLETED.

    A function to test spectral clustering sensitivity to parameter choice.
    """
    # the number of samples to generate
    num_samples = 500

    """
    Choose parameters
    """
    var = 1.0  # exponential_euclidean's sigma^2
    laplacian_normalization = 'unn'
    chosen_eig_indices = None

    """
    Choose candidate parameters
    """
    parameter_type_eps = 'eps' 
    parameter_type_k = 'k'
    parameter_candidate_eps = np.arange(0,1,0.05)  # the number of neighbours for the graph (if 'k')
    parameter_candidate_k = np.arange(1,102,5)                           # or the epsilon threshold (if 'eps')
    parameter_performance_eps = []
    parameter_performance_k = []

    for parameter in tqdm.tqdm(parameter_candidate_eps):
        # Generate data
        X, Y = two_moons(num_samples, 1, 0.02)
        num_classes = len(np.unique(Y))
        
        parameter_wrap = {parameter_type_eps: parameter}
        
        W = build_similarity_graph(X, **parameter_wrap)
        L = build_laplacian(W, laplacian_normalization)

        Y_rec = spectral_clustering(L, chosen_eig_indices, num_classes)

        parameter_performance_eps += [skm.adjusted_rand_score(Y, Y_rec)]
        
    for parameter in tqdm.tqdm(parameter_candidate_k):
        # Generate data
        X, Y = two_moons(num_samples, 1, 0.02)
        num_classes = len(np.unique(Y))
        
        parameter_wrap = {parameter_type_k: parameter}
        
        W = build_similarity_graph(X, **parameter_wrap)
        L = build_laplacian(W, laplacian_normalization)

        Y_rec = spectral_clustering(L, chosen_eig_indices, num_classes)

        parameter_performance_k += [skm.adjusted_rand_score(Y, Y_rec)]
        
    plt.figure()
    plt.plot(parameter_candidate_eps, parameter_performance_eps)
    plt.title(f'parameter sensitivity ({parameter_type_eps})')
    plt.figure()
    plt.plot(parameter_candidate_k, parameter_performance_k)
    plt.title(f'parameter sensitivity ({parameter_type_k})')
    plt.show()
```

```python
parameter_sensitivity()
```

### Q1.2: While varying the parameter of `worst_case_blob`, try to select a value of $\epsilon$ that keeps the graph connected (the function `min_span_tree` can help you). What happens? Can you tell when it is a bad idea to use this approach?

The $\epsilon$ parameter that we need to give to our function so that the graph is connected must be smaller than the distance of the outlier to the nearest point. This distance is about $e^{- \frac{\delta^{2}}{2var}}$ therefore epsilon decreases exponentially with $\delta^{2}$. This $\epsilon$ approach tends to remove outliers from the graph by casting it out of the main connected component. If we want to build a connected graph then we need to take $\epsilon$ very small and accept all weights, therefore making the graph very dense, which is a bad idea as we want a sparse graph. 

Let's use the function given in the hint to make it connected :

### Q1.3: Is it important to build a connected graph? Compare k-NN to $\epsilon$-graphs with respect to the difficulty of building a connected graph.

In general it is important to build a connected graph. Indeed, if we assume our features represent one phenomenon, then having different components in the graph removes the belonging of our features to this single phenomenon. If there exists two distincts components in the graph, then these two components will immediatly define two clusters. If, for instance, there are more disconnected components than the true number of clusters, the algorithm will not be able to produce a good result. So if we want to find 2 clusters, we will immediately have the output. With k-NN, building a connected component is easier if there are some outliers in the data and the data is uniformly spaced, as every point is at least connected to another if $k>0$. The $\epsilon$ approach is a bad idea in this case as shown in question 1.2. On the other hand, if points are clustered, the k-nn approach connect together the points of each cluster and creates several connected components. Taking $\epsilon$ bigger than the distance between any two clusters can then create a connected graph. Let's illustrate these arguments :

### Q2.1: Build a graph starting from the data generated in `blobs`, and keep the graph connected. Use `spectral_clustering` to cluster the data. Motivate your choice on which eigenvectors to use and how you computed the clustering assignments from the eigenvectors. Now compute a similar clustering using the built-in k-means and compare the results.

We have seen in class that the smallest eigenvalue is 0 and that its multiplicity equals the number of connected components. Its eigenspace is spanned with the indication vectors of the connected components. If we have only one connected component, the only eigenvector of 0 is constant and therefore using this eigenvector is not useful for clustering. In this case however, the other eigenvectors are meaningful for the clustering. As we have seen in class, the Rayleigh-Ritz theorem states that the eigen vector associated with the second smallest (non zero) eigen value of the Laplacian matrix is a minimizer of the relaxed version of the MinCut. 

In order to cluster points in 2 groups, a method is to apply the k_means $(k = 2)$ to this vector exactly like slide 34 of the course suggests.

### Q2.3: In the function spectral_clustering, implement an automatic strategy to choose the eigenvectors to use for clustering, when chosen_eig_indices=None. Explain your strategy. Generate a dataset with 4 blobs and test your function. What happens when the variance of the blobs increase?

If we are looking for more than 2 clusters, Ulrike von Luxburg proved in the paper "A tutorial on Spectral Clustering" https://arxiv.org/abs/0711.0189, that for the MinCut problem with $k \geq 3$ clusters : 

$$ min_{A_1, A_2, ..., A_k} cut(A_1, A_2, ..., A_k) = min_{A_1, A_2, ..., A_k} \sum_{k = 1}^kW(A_i, \bar A_i)$$

Where : 
$$W(A, B) := \sum_{i \in A, j \in B} w_{ij}, \text{ and } \bar A \text{ is the complement of } A $$

Which is equivalent to:
$$
\begin{alignat}{2}
&\!\min_{A_1, A_2, ..., A_k}       &\qquad&\text{Tr}(H^TLH)\\
&\text{subject to} &      & H^TH = I\\
&                  &      & H_{ik} = \begin{cases}
                                                1 &  \text{if }V_i \in A_k \\
                                               -1 &  \text{if }V_i \in \bar A_k
                                     \end{cases}
\end{alignat}
$$

The relaxed version of this problem is :

$$
\begin{alignat}{2}
&\!\min_{A_1, A_2, ..., A_k}       &\qquad&\text{Tr}(H^TLH)\\
&\text{subject to} &      & H^TH = I\\
&                  &      & H \text{is a matrix} \in \mathbb{R}^{n, k}\\
\end{alignat}
$$

And the solution of the relaxed version can still be found using the Rayleigh-Ritz theorem, with H being the matrix containing the first k eigenvectors of L as columns in an orthogonal basis.
To get an approximate solution for the non relaxed version of the MinCut we then apply the k-means algorithm on the rows of the first k eigenvectors stacked as colomns. This solution is implemented in the algorithm spectral_clustering. One particularity is that if the graph is not connected, we take as first eigenvector the eigenvectors of 0, whereas if the graph is connected we take as first eigenvector the eigenvector of the smallest non negative eigenvalue.

The Ncut minimization problem separates the clusters by weighting the cuts with their volume (and thus the sum of weights of each cluster). The RatioCut minimization problem separates the clusters by weighting the cuts with the cardinal of each cluster. As we have seen in class, the random-walk laplacian is used to minimize the NCut problem whereas the unnormalized Laplacian is used to minimize the RatioCut problem. This explains why in the second plot that the number of samples in each clusters are almost equivalent, but the volume of each cluster are lower (some points are far from each other), which leads to a bad clustering due to the structure of the data, with one of the two clusters having points far from each other. In the second plot however, the volumes of the green circle and the red one are high (because the points are closed to each other), which leads to a good clustering.

### Q2.8: If we did not have access to "true" labels, how could we evaluate the clustering result (or what should we not use as evaluation)?

If we dont have the true labels, the performance of the clustering algorithm cannot be measured with external measures as the ARI (or the Adjusted Mutual Information for example).
We have to use internal measures instead, i.e. that only depends on the clustering result. Something that we shouldn't do on the other hand is to use only within-cluster criteria, as it only consider that the points of a cluster are closed to their centroid.

An internal measure used in graph theory is the modularity.

### Q3.2: A full graph built between the pixels of a 50  50 image corresponds to $50^2$ nodes. Solving the full eigenvalue problem in this case would scale in the order of $2^{34}$ . Even on weak hardware this takes only seconds to minutes. Segmenting a Full HD picture of 1920  1080 would scale in the order of $2^{64}$ (about a month on a decent machine). Beyond that, the large picture would require to store in memory a graph over millions of nodes. A full graph on that scale requires about 1TB of memory. Can you think two simple techniques to reduce the computational and occupational cost of Spectral Clustering?

1. We can use eigenvalues algorithm to compute only the eigenvectors associated with the $m$ lowest (or highest) eigenvalues. This allows us to reduce considerably the computational cost since we don't have to do the computation for every eigenvectors.

2. We can downsample the image before feeding it to the spectral clustering algorithm.

3. We can convolve the image with a kernel and do maxpooling to reduce the resolution of the image and therefore reduce its size.
